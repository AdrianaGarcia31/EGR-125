```python
import pandas as pd
```

```python
#The index column make easy to look at values
weather = pd.read_csv('weather_data.csv',index_col = "DATE")
```

```python
weather.head()
```

```python
#This code will count the number of missing values in each column. 
#apply() applies the isnull function to each element on the Data Frame

weather.apply(pd.isnull).sum()
```

```python
weather.columns
```

```python
len(weather.columns)
```

```python
#shape[0] gives the number of rows and shape[1] gives the number of columns
#To know which columns have missing values in percentage
weather.apply(pd.isnull).sum()/weather.shape[0]
```

```python
#Using the data documentation and from analyzing the percentage. I will choose some of them for my new weather data
new_weather = weather[["PRCP","SNOW","SNWD","TMAX","TMIN"]].copy()
```

```python
#Now we can see again the missing values for these 6 information
new_weather.apply(pd.isnull).sum()/new_weather.shape[0]
```
